```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
```

```python
X = housing.data
y = housing.target
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score
```

```python
lr = LinearRegression()
dt = DecisionTreeRegressor()
svr = SVR()
```

```python
estimators = [('lr',lr),('dt',dt),('svr',svr)]
```

```python
import numpy as np
for estimators_item in estimators:
  scores = cross_val_score(estimators_item[1], X,y,scoring='r2', cv=10)
  print(estimators_item[0],np.round(np.mean(scores),2))
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
vr = VotingRegressor(estimators=estimators)
scores = cross_val_score(vr, X,y,scoring='r2',cv=10)
print("voting Regressor", np.round(np.mean(scores),2))
```

```python
for i in range(1,4):
  for j in range(1,4):
    for k in range(1,4):
      vr = VotingRegressor(estimators=estimators, weights=[i,j,k])
      scores = cross_val_score(vr, X, y, scoring='r2',cv=10)
      print("for i={} j={} k={}".format(i,j,k), np.round(np.mean(scores),2))
```

```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X_new, y_new = make_regression(n_samples=10000, n_features=5,n_targets=1)
from sklearn.tree import DecisionTreeRegressor
```

```python
dr1 =  DecisionTreeRegressor(max_depth=1)
dr2 =  DecisionTreeRegressor(max_depth=2)
dr3 =  DecisionTreeRegressor(max_depth=3)
dr4 =  DecisionTreeRegressor(max_depth=4)
dr5 =  DecisionTreeRegressor(max_depth=None)
```

```python
estimators = [('dr1',dr1),('dr2',dr2),('dr3',dr3),('dr4',dr4),('dr5',dr5)]
```

```python
for estimators_item in estimators:
  scores = cross_val_score(estimators_item[1],X_new,y_new,scoring='r2',cv=10)
  print(estimators_item[0],np.round(np.mean(scores),2))
```

```python
vr = VotingRegressor(estimators=estimators)
```

```python
scores = cross_val_score(vr,X_new, y_new, scoring='r2',cv=10)
print("Voting Regressor:",np.round(np.mean(scores),2))
```
