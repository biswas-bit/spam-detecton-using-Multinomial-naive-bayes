```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
X , y= make_blobs(
    n_samples=300,
    n_features=2,
    centers=4,
    cluster_std=0.6,
    random_state=42
)
```

```python
plt.scatter(X[:,0], X[:,1])
plt.title("synthetic Clustering dataset")
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,11):
  km = KMeans(n_clusters=i)
  km.fit_predict(X)
  wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
```
