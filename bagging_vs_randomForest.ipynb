```python
from sklearn.datasets import make_classification
import pandas as pd
```

```python
X,y = make_classification(n_samples=100, n_features=6,n_classes=2)
```

```python
df = pd.DataFrame(X,columns=['col1','col2','col3','col4','col5','col6'])
df["target"]=y
df.head()
```

```python
from  sklearn.ensemble import  BaggingClassifier,RandomForestClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
bag = BaggingClassifier(max_features=3)
bag.fit(df.drop('target',axis=1),df["target"])
```

```python
plt.figure(figsize=(12,12))
plot_tree(bag.estimators_[0])
plt.show()
```

```python
rf = RandomForestClassifier()
rf.fit(df.drop('target',axis=1),df["target"])
```

```python
plt.figure(figsize=(12,12))
plot_tree(rf.estimators_[0])
plt.show()
```
